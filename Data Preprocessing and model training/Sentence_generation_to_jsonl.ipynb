conversion to jsonl

```python
def csv_to_jsonl(csv_file, jsonl_file):
    df = pd.read_csv(csv_file)
    df['sentence'] = df.apply(generate_sentences, axis=1)

    with open(jsonl_file, "w") as outfile:
        for index, row in df.iterrows():
            sentence = row["sentence"]
            annotations = []

            def add_annotation(entity, label):
                if isinstance(entity, str):
                    start = sentence.find(entity)
                    while start != -1:
                        end = start + len(entity)
                        annotations.append({"start": start, "end": end, "label": label})
                        start = sentence.find(entity, end)

            add_annotation(row["Area Name"], "TOWN")
            add_annotation(row["SUB-DISTRICT NAME"], "SUBDISTRICT")
            add_annotation(row["DISTRICT NAME"], "DISTRICT")
            add_annotation(row["STATE/UT NAME"], "STATE")

            if annotations:
                json_data = {"text": sentence, "spans": annotations}
                json.dump(json_data, outfile)
                outfile.write("\n")
```
